```python
import numpy as np
import math
from IPython.display import clear_output
```

```python
A =  (500 * 5.5 / 4.5) #+ math.sqrt((5.5 / 3.5) * (500 / 4.5) ** 2)
l = 1 / 100  #(100 - math.sqrt(100))
r = 1  * (l * A)
start = 10
quantity = 500
ready_clients = []
clients = []
clients_start = []
T = 0
```

```python
for i in range (start):
    A = int((np.random.pareto(5.5) + 1) * 500)
    r_med = int(np.random.uniform(r / 5.5, r * (10 / 5.5)))
    Rk = int(np.random.uniform(0,A))
    T0 = 0
    client = [T0,A,Rk,r_med]
    clients_start.append(client)

T0 = 0
for i in range (quantity):
    A = int((np.random.pareto(5.5) + 1) * 500)
    r_med = int(np.random.uniform(r / 5.5, r * (10 / 5.5)))
    T0 += int(np.random.exponential(11))
    Rk = 0
    client = [T0,A,Rk,r_med]
    clients.append(client)
    
```

```python
def condition(number,rk,Rk,r_med,Tk,Tk0):
    if number == 1 :
        return rk
    if number == 2 :
        return Tk0
    if number == 3 :
        return rk*Tk/Rk
    if number == 4 :
        return rk/Rk
```

```python
A_last_sum = sum([c[1] for c in clients_start])
while len(clients) != 0:
    T += 1
    try:
        if clients[0][0]  <= T :
                clients_start.append(clients[0])
                del clients[0]
    except:
        pass
            
    test = 0
    x = 0
    A_sum = 0
    R_sum = 0
    for i in range (0,len(clients_start)):
        
        Tk0 = clients_start[i][0]
        Tk = T - Tk0
        r_med = clients_start[i][3]
        rk = int(np.random.uniform(r_med * 0.7, r_med * 1.3))
        Rk = clients_start[i][2]+0.000000000000001
        A_sum+=clients_start[i][1]
        R_sum+= Rk
        
        # 1 = argmax rk(t) 
        # 2 = argmin Tk0 
        # 3 = argmax rk(t)*(t-Tk0)/Rk(t)
        # 4 = argmax rk(t)/Rk(t)
        arg = condition(3,rk,Rk,r_med,Tk,Tk0)
        if arg > test :
            test =arg
            x = i
    if len(clients_start)>0:
        rx = clients_start[x][3]
        Rx = clients_start[x][2] 
        Ax = clients_start[x][1]
        Rx += int(np.random.uniform(rx * 0.7, rx * 1.3))
        clients_start[x][2] += Rx
        if Rx >= Ax:
            clients_start[x][0] = T - clients_start[x][0]
            ready_clients.append(clients_start[x])
            del clients_start[x]
    print(len(clients),len(clients_start),len(ready_clients),T)
#     print(A_sum-A_last_sum,R_sum)
    A_last_sum = A_sum

```
